```python
# label이 포함된 시간 return하는 함수
def stack_time(df, *labels):    # df = idf,  Labels = 입력쿼리문 토큰 ex) '자동차'
    names = ['person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train', 'truck', 'boat', 'traffic light',
        'fire hydrant', 'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow',
        'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag', 'tie', 'suitcase', 'frisbee',
        'skis', 'snowboard', 'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard', 'surfboard',
        'tennis racket', 'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana', 'apple',
        'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch',
        'potted plant', 'bed', 'dining table', 'toilet', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
        'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock', 'vase', 'scissors', 'teddy bear',
        'hair drier', 'toothbrush']
    times = []
    for label in labels :
        if type(label) == int:
            over = 0
        else :
            label, over = label
            over -= 1
        times.append(list(zip(df[df['label'] == str(label)+'_'+str(over)]['S_time'].values, df[df['label'] == str(label)+'_'+str(over)]['E_time'].values)))
        print(f'searching for {over+1} {names[label]}')
    return time
```

```python
# 시간들 중 교집합 되는 시간을 return하는 함수
def is_everything(times) :
    while len(times) > 1 :
        c1 = times.pop()
        c2 = times.pop()
        data = []
        c1.sort()
        c2.sort()
        for time1 in c1:
            for time2 in c2:
                if time2[0] > time1[1]:
                    break
                elif time1[0] > time2[1]:
                    continue
                else : 
                    data.append((max(time1[0], time2[0]), min(time1[1],time2[1])))
        for time1 in c2:
            for time2 in c1:
                if time2[0] > time1[1]:
                    break
                elif time1[0] > time2[1]:
                    continue
                else : 
                    data.append((max(time1[0], time2[0]), min(time1[1],time2[1])))

        data = list(set(data))
        data.sort()
        times.append(data)
    times = list(times[0])
    times.sort()
    sol = []
    for i in range(len(times)):
        sol.append(times[i])
#        if (times[i][1] - times[i][0]) >= 1:
#            sol.append(times[i])
    return sol
```

```python
# 시간들 중 합집합 되는 시간을 return하는 함수
def is_anything(times) :
    times = sum(times, [])
    times.sort()
    sol = []
    time = list(times[0])
    for i in range(len(times)):
        if times[i][0] > time[1]:
            sol.append(tuple(time))
            time = list(times[i])
        elif times[i][1] > time[1]:
            time[1] = times[i][1]
    sol.append(tuple(time))    
    return sol
```

```python
# 음성 검색 함수
def make_function(proba_df):
    t_df = proba_df.sort_values('sec')
    song_voice = t_df[t_df['pred_voice'].str.contains('human_voice|song')][['song', 'human_voice','pred_voice']].groupby(['song', 'human_voice']).count()
    z = np.polyfit(song_voice.reset_index()['song'], song_voice.reset_index()['human_voice'], 1)
    song_voice_f = np.poly1d(z)
    song_others = t_df[~t_df['pred_voice'].str.contains("human_voice")][['song', 'others','pred_voice']].groupby(['song', 'others']).count()
    z = np.polyfit(song_others.reset_index()['others'], song_others.reset_index()['song'], 1)
    others_song_f = np.poly1d(z)
    return song_voice_f, others_song_f

def create_labels(merged_df, song_voice_f, others_song_f):
    result_dict = dict(zip(['women','men','song'],[[] for i in range(3)]))
    song_voice_mean = (merged_df['song']+merged_df['human_voice']).mean()
    if np.argmax(merged_df.describe().loc['mean'][:5]) == 1: #human_voice 위주의 영상
        print("Voice가 주요 요소인 동영상 입니다.")
        voice_baseline = np.percentile(merged_df['human_voice'], 15)
        song_baseline = np.percentile(merged_df['song'], 50)
    elif np.argmax(merged_df.describe().loc['mean'][:5]) == 4: #song 위주의 영상
        print("Song이 주요 요소인 동영상 입니다.")
        voice_baseline = np.percentile(merged_df['human_voice'], 50)
        song_baseline = np.percentile(merged_df['song'], 25)
    else:
        voice_baseline = np.percentile(merged_df['human_voice'], 25)
        song_baseline = np.percentile(merged_df['song'], 25)

    print("song 베이스라인: ",song_baseline, "/ 목소리 베이스라인: ", voice_baseline)
    for i in range(len(merged_df)):
        voice_pct = merged_df['human_voice'].iloc[i]
        song_pct = merged_df['song'].iloc[i]
        others_pct = merged_df['others'].iloc[i]
        pred_voice = merged_df['pred_voice'].iloc[i]
        
        if pred_voice=='human_voice' or ((voice_pct >= song_voice_f(song_pct)) and (voice_pct>=voice_baseline)):
            pred_gender = merged_df['pred_gender'].iloc[i]
            if pred_gender == 'read_women':
                result_dict['women'].append(1)
                result_dict['men'].append(0)
                if song_pct>song_baseline:
                    result_dict['song'].append(1)
                else:
                    result_dict['song'].append(0)
            else:
                result_dict['women'].append(0)
                result_dict['men'].append(1)
                if song_pct>song_baseline:
                    result_dict['song'].append(1)
                else:
                    result_dict['song'].append(0)
        elif pred_voice=='song' or (song_pct>=song_baseline and (song_pct >= others_song_f(others_pct))):
            result_dict['song'].append(1)
            if (voice_pct>=voice_baseline) and (pred_gender == 'read_women'):
                result_dict['women'].append(1)
                result_dict['men'].append(0)
            elif (voice_pct>=voice_baseline) and (pred_gender == 'read_men'):
                result_dict['women'].append(0)
                result_dict['men'].append(1)
            else:
                result_dict['women'].append(0)
                result_dict['men'].append(0)
        else:
            result_dict['women'].append(0)
            result_dict['men'].append(0)
            result_dict['song'].append(0)
    result_df = pd.DataFrame(result_dict, index=merged_df['sec'])
    return result_df
```

```python
# 영상 검색 결과 내 연속된 음원 검색

def time_index(times):
    time_ls = []
    [time_ls.extend([round(f,1) for f in np.arange(time[0],time[1],0.1).tolist()]) for time in times]
    return time_ls

def queue_sound(result_df, times, s_selector, min_sec=0.5, duration=0):
    # min_sec은 최소 유지 시간 (최소 0.5초가 지속되지 않는 시간대는 탈락)
    if len(times) == 0:
        return [] 
    time_idx = time_index(times)
    try:
        test = result_df.loc[time_idx]

        ls0 =test[s_selector].tolist()
        final_ls = np.array(ls0)
        if duration : 
            for time in range(1, int(duration*10+1)): 
                try : 
                    tmp_ls = [0] * time
                    tmp_ls.extend(ls0[:-time])
                    final_ls += np.array(tmp_ls)
                except : 
                    break

        tmp_df = pd.DataFrame(final_ls, index=test.index)
        queue = (tmp_df[tmp_df[0]>=1].index*10).tolist()
        if len(queue):
            packet, tmp = [], []
            v = queue.pop(0)
            tmp.append(v)

            while(len(queue)>0):
                vv = queue.pop(0)
                if v+1 == vv:
                    tmp.append(vv)
                    v = vv
                else:
                    packet.append(tmp)
                    tmp = []
                    tmp.append(vv)
                    v = vv
            packet.append(tmp)
            result = [(p[0]/10,p[-1]/10) for p in packet if len(p)>min_sec*10]
            return result
    except:
        print("적합한 음원이 없습니다.")
```

```python
def strtoint(text) : 
    dic = {'하나' : '1/NR', '둘':'2/NR', '셋':'3/NR', '넷':'4/NR', '다섯': '5/NR', '여섯': '6/NR', '일곱':'7/NR', '여덟':'8/NR', '아홉':'9/NR', '열': '10/NR','한' : '1/NR', '두':'2/NR', '세':'3/NR', '네':'4/NR'}
    for i in range(len(text)) : 
        if i > len(text):
            break
        for j in range(len(text[i])):
            if (len(text[i]) == 1) and text[i][0] == '명의/NNG':
                text[i] = []
                break

            if text[i][j].split('/')[1] == 'MDN' or text[i][j].split('/')[1] == 'NR' or text[i][j].split('/')[1] == 'NNG':
                try : 
                    if (len(text[i+1]) == 1) and text[i+1][0] == '명의/NNG':
                        text[i].extend(text[i+1])
                except :
                    pass
                try :
                    text[i][j] = dic[text[i][j].split('/')[0]]
                except : 
                    pass 
    new_text = ''
    for txt in text:
        new_text += ' '
        for t in txt:
            new_text += t.split('/')[0]         
    new_text
    datas = kkm.pos(new_text,flatten=False, join=True) #kkma
    return datas
```

```python
def scene_search(i_df, s_df, o_url, y_url, min_sec=1.0, duration=0.4):
    kkm = Kkma()
    i_text = input('영상 검색 : ')
    datas = kkm.pos(i_text,flatten=False, join=True) #kkma
    
    try : 
        num = datas.index([data for data in datas if ('소리/NNG' in data) or ('목소리/NNG' in data) or ('말하/VV' in data) or ('부르/VV' in data) or ('들리/VV' in data) or ('나오/VV' in data)][0])
    except : 
        pass

    try : 
        if ('여성/NNG' in datas[num-1]) or ('여자/NNG' in datas[num-1]):
            print('searching women voice')
            s_selector = 'women'
            del datas[num]       # 소리가 포함된 어절을 날리고
            del datas[num-1]    # '여성/여자' 처럼 식별 확인된 어절도 날린다.
        elif ('여성/NNG' in datas[num]) or ('여자/NNG' in datas[num]):
            print('searching women voice')
            s_selector = 'women'
            del datas[num]      # 소리+여성/여자가 한 어절에 있어서 그 하나를 날린다.

        elif ('남성/NNG' in datas[num-1]) or ('남자/NNG' in datas[num-1]):
            print('searching men voice')
            s_selector = 'men'
            del datas[num]
            del datas[num-1]
        elif ('남성/NNG' in datas[num]) or ('남자/NNG' in datas[num]):
            print('searching men voice')
            s_selector = 'men'
            del datas[num]

        elif ('노래/NNG' in datas[num-1]) or ('음악/NNG' in datas[num-1]) or ('노랫소리/NNG' in datas[num-1]) :
            print('searching song sound')
            s_selector = 'song'
            del datas[num]
            del datas[num-1]
        elif ('노래/NNG' in datas[num]) or ('음악/NNG' in datas[num]) or ('노랫소리/NNG' in datas[num]):
            print('searching song sound')
            s_selector = 'song'
            del datas[num]

        elif ('사람/NNG' in datas[num-1]) or ('목소리/NNG' in datas[num-1]) :
            print('searching human voice')
            s_selector = 'human'
            del datas[num]
            del datas[num-1]
        elif ('사람/NNG' in datas[num]) or ('목소리/NNG' in datas[num]):
            print('searching human voice')
            s_selector = 'human'
            del datas[num]
            
        else : 
            s_selector = 0
    except:
        s_selector = 0


    times = []

    for i in range(len(datas)):
        for data in datas[i]:
            if data.split('/')[1] == 'NNG':
                try : 
                    times.append(is_anything(stack_time(i_df, *dic[data.split('/')[0]])))
                except :
                    pass
            elif data.split('/')[1] == 'NR':
                if datas[i][-1].split('/')[1] == 'JKG':
                    try :
                        item = dic[[v.split('/')[0] for v in datas[i+1] if v.split('/')[1] == 'NNG'][0]]
                        times.append(is_anything(stack_time(i_df,*list(zip(item, [int(data.split('/')[0])] * len(item))))))
                    except :
                        pass
                elif (i > 0) and 'NNG' in [s.split('/')[1] for s in datas[i-1]] :
                    try :
                        num = [s.split('/')[1] for s in datas[i-1]].index('NNG')
                        item = dic[[s.split('/')[0] for s in datas[i-1]][num]]
                        times.append(is_anything(stack_time(i_df,*list(zip(item, [int(data.split('/')[0])] * len(item))))))
                    except :
                        pass

    if len(times):
        times = is_everything(times)
    elif s_selector != 0: 
        times = [(s_df['sec'].values[0], s_df['sec'].values[-1])]

    # 음성 디텍션 코드

    if s_selector == 0:
        pass
    else : 
        song_voice_f, others_song_f = make_function(s_df)
        result_df = create_labels(s_df, song_voice_f, others_song_f)
        result_df['human'] = result_df['men'] | result_df['women']
        times = queue_sound(result_df, times, s_selector, min_sec=min_sec, duration=duration)

    for data in times :
        print(f'{data}, {o_url}{int(data[0])}s')
        print(f'yolo :  {y_url}{int(data[0])}s')
```
